```python
s = """| СИНОНИМИ = [[rt]]
| СРОДНИ ДУМИ = [[конфронтирам]]
| ПРОИЗВОДНИ ДУМИ = 
| ДРУГИ = """
```

```python
patt = r"(?:СИНОНИМИ =\s+([^|]*))"
match = re.search(patt, s, re.M)
raw_syn = match.group(1)
re.findall(r"\[\[(.*?)\]\]", raw_syn, re.M)
```

```python
from SPARQLWrapper import SPARQLWrapper, TSV, CSV

sparql = SPARQLWrapper("http://dbpedia.org/sparql")
query = """
PREFIX dbo: <http://dbpedia.org/ontology/>
PREFIX foaf:  <http://xmlns.com/foaf/0.1/>

SELECT count(*)
WHERE
{
   ?person a dbo:Person .
}

"""
sparql.setQuery(query)
sparql.setReturnFormat(CSV)
results = sparql.query().convert()
```

```python
import pandas as pd
```

```python
path = "/home/karimlulu/Documents/checklist_stat.txt"
df = pd.read_csv(path)
```

```python
df
```

```python
mapp = {1: 'Need it',
        2: 'Have it',
        3: 'Not for me'}
g_map = {1:"man", 139:"woman"}
```

```python
df["state"] = df["state"].map(lambda x: mapp.get(x, x))
```

```python
df["item_gender"] = df["item_gender"].map(lambda x : g_map.get(x, x))
```

```python
df.head()
```

```python
pct = pd.pivot_table(df, index=["name", "item_gender"], columns="state", values=["count"]).reset_index()
print(pct.head())
pct.columns = pct.columns.droplevel()
```
